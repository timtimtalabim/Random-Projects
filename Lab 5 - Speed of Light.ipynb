```python
import numpy as np
import matplotlib.pyplot as plt

#distance is in meteres
#time is in nano seconds


#wavelength 525 color green

Distance_1 = [2.06, 3.00, 3.50, 4.00, 4.50]
Time_1 = [8, 12.5, 14, 15.5, 16.5]


#wavelength 632 color red

Distance_2 = [1.48, 2.036, 2.502, 3.00, 3.50, 4.00]
Time_2 = [7, 7.8, 9.6, 11.2, 12.4, 15]

#wavelength 394 color purple

Distance_3 = [1.00, 1.50, 2.00, 2.50, 3.00]
Time_3 = [4, 5, 7, 8.5, 10]

#wavelength 465 color blue

Distance_4 = [1.00, 1.50, 2.00, 2.50, 3.00, 3.50]
Time_4 = [4, 6, 8, 9, 11.5, 13.5]

#wavelength 405 color purple

Distance_5 = [1.00, 1.50, 2.00, 2.50, 3.00, 3.50]
Time_5 = [5, 6, 8, 10.5, 13, 15]

#wavelength 850 infared 

Distance_6 = [1.00, 1.50, 2.00, 2.50, 3.00, 3.50, 4.00, 4.50, 5.00, 5.50]
Time_6 = [4.5, 6.5, 8, 9.5, 12.5, 13.5, 15, 16.5, 18, 20.5]

Distances = [Distance_1, Distance_2, Distance_3, Distance_4, Distance_5, Distance_6]
Times = [Time_1, Time_2, Time_3, Time_4, Time_5, Time_6]

Times = [[i*10**-9 for i in j] for j in Times]

slopes = []
colors = ['g', 'r', 'tab:purple', 'b', 'xkcd:purple', 'tab:gray']

for i in range(0,6):
    m, b = np.polyfit(Times[i], Distances[i], 1)
    slopes.append(m)

print(slopes)
final = sum(slopes)/6
std = np.std(slopes)

print("c value:", final, "One std dev:", final-std, "-", final+std)

plt.figure(figsize=(5,3), dpi = 200)
x = np.linspace(0, 20)
for i in range(0, 6):
    y = [slopes[i]*j for j in x]
    plt.plot(x, y, label="{:.2e}".format(slopes[i]), color=colors[i])
    
y = [final*j for j in x]
plt.plot(x, y, label=(f'avg: {"{:.2e}".format(final)}'), color='y')
plt.legend(fontsize='small')
plt.xlim(0, 20)
plt.ylim(0)
plt.ylabel("Velocity (m/s)")
plt.xlabel("Time (ns)")
```
