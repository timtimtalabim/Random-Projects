```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Main Function change U for potential energy and change L for width
def solveEquation(U = 400, L = 0.1):
    
    # Creates x points for the graph
    points = np.linspace(0,U,U*1000)
    
    # Creates the left hand side and the right hand side of the equation respectively
    f = lambda x: (2 * (U*x - x*x)**0.5)
    g = lambda x: (2 * x - U)*np.tan((26.3*x * L * L)**0.5)
    
    # Puts answers into a list and prints them
    answers = [round(x) for x in points if(f(x) < g(x) + 0.05 and f(x) > g(x) - 0.05)]
    answers = list(dict.fromkeys(answers))
    print(answers)
    
    # Plots out the graphs and points of intersection
    plt.figure(figsize=[12, 8])
    plt.plot(points,f(points), label='left hand side')
    plt.plot(points,g(points), label='right hand side')
    for x in answers:
        plt.scatter(x, f(x), label = (x, f(x)))
    plt.xlim(0,U)
    plt.ylim(-1000,1000)
    plt.title('Graph')
    plt.legend(fontsize=12)
    plt.show()
    
    # Plots out the potential energy well
    plt.figure(figsize=[4, 8])
    n = 4
    for x in answers:
        plt.axhline(x, linestyle='-', label = ('n = %s, E = %s'%(n, answers[n])))
        n -= 1
    plt.title('Potential Energy Well', size = 12)
    plt.legend(loc=(1.04,0))
    plt.show()
            
solveEquation()
```
