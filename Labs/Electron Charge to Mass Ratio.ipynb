```python
import matplotlib.pyplot as plt
import numpy as np

e_by_m_acc = 1.7588196 * 10**11
B = 7.8*10**-4
v = 150

amps = [1, 1.066, 1.133, 1.2, 1.266, 1.333, 1.4, 1.466, 1.533, 1.6, 1.666, 1.733, 1.8, 1.866, 1.933]

radius = [4.375, 4.075, 3.85, 3.725, 3.575, 3.225, 2.925, 2.75, 2.5, 2.35, 2, 1.9, 1.775, 1.675, 1.55]

a = [(1/i)**2 for i in amps]
r = [(i*10**-2)**2 for i in radius]

m, b = np.polyfit(r, a, 1)

e_by_m_calc = (2*v/(B**2))*m

error = (e_by_m_calc - e_by_m_acc)/e_by_m_acc

print("Calculated e/m value:", "{:e}".format(e_by_m_calc), "C/kg")
print("Error:", error*100, "%")

f = lambda x: m*x + b
y = [f(i) for i in r]

plt.plot(r, a, 'o', label='Data')
plt.plot(r, y, '-', label='Linear Regression Line')
plt.xlabel('Radius Squared $(m^2)$')
plt.ylabel('1/Current Squared $(A^2)$')
plt.legend()
plt.title('Trial #1')
plt.show()
```
